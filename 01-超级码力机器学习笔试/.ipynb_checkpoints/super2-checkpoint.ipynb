###### A->label :   P(A,label)/P(A)=num(A,label)/num(A)  > CONFIDENCE

```python
SUPPORT = 0.1
CONFIDENCE = 0.7 
```

Apriori算法原理总结: https://www.cnblogs.com/pinard/p/6293298.html

Apriori算法： 空间复杂度低，时间复杂度高
FP Growth算法：空间复杂度高，时间复杂度低

# Apriori思路大体分三步：
### 1.观察数据，对数据预处理
### 2.对于A->Label: 根据支持度和置信度条件删除dataframe中一部分特征
### 3.对于AB->Label: 根据支持度和置信度条件，直接得出结果

## 观察数据，观察数据，对数据预处理：

## 对于A->Label:  根据支持度和置信度条件删除一部分特征

```python
list_dfcolname = df.columns.values.tolist()
print('n_SUPPORT:', df.shape[0] * SUPPORT)
print()
n_A = df['Label'].groupby(df['VTYPE_1']).count()
print(n_A)
# print(1 in n_A.index)
# n_A = df['Label'].groupby(df['VTYPE_5']).count()
# print(1 in n_A.index)
for i in range(df.shape[1]):
    n_A = df['Label'].groupby(df[list_dfcolname[i]]).count()
    n_A_and_label = df['Label'].groupby([df['Label'], df[list_dfcolname[i]]]).count()
    if 1 in n_A.index:
        #对于A->Label:  根据支持度和置信度条件删除一部分特征
        if n_A[1] <= df.shape[0] * SUPPORT:
            df.drop([list_dfcolname[i]],axis=1,inplace=True)
        else:
            if 2 in (n_A_and_label.index.labels[0] + n_A_and_label.index.labels[1]): #是否存在Label为1，A也为1
                if n_A_and_label[1][1]/n_A[1] <= CONFIDENCE:#num(A,label)/num(A) < CONFIDENCE:
                    df.drop([list_dfcolname[i]],axis=1,inplace=True)
    else:
        #特征全为零，删除
        df.drop([list_dfcolname[i]],axis=1,inplace=True)
print(df.shape)                  
```

## 接下来找出规则左侧两项的情况，交叉组合后总的判断次数为 32的阶乘：32！

```python
print(df.columns.values.tolist()[:-1])
#组合AB特征
list_A_B = []
for A in df.columns.values.tolist()[:-1]:
    for B in df.columns.values.tolist()[:-1]:
        n_AB = df['Label'].groupby([df[A],df[B]]).count()
        n_AB_and_label = df['Label'].groupby([df['Label'],df[A],df[B]]).count()
        
        if 2 in (n_AB.index.labels[0] + n_AB.index.labels[1]): #是否存在A为1，B也为1
            if n_AB[1][1] > df.shape[0] * SUPPORT:  #AB满足支持度
                 #是否存在Label为1，A也为1, B也为1， 即A,B,Label共存
                if 3 in (n_AB_and_label.index.labels[0] + n_AB_and_label.index.labels[1] +n_AB_and_label.index.labels[2]): 
                    if n_AB_and_label[1][1][1]/n_AB[1][1] > CONFIDENCE:#num(A,B,label)/num(A,B) > CONFIDENCE(满足置信度)
                        if A == B:
                            list_A_B.append(A) 
                        else:
                            list_A_B.append({A,B})    

list_A_B                      
# print(n_AB_and_label.index)
# n_AB_and_label  
```

## 列表去重便是最终结果了

```python
#list_A_B=list(set(list_A_B))

result=[]
for i in list_A_B:
    if i not in result:
        result.append(i)
result
```
